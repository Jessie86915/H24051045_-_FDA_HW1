## 5. Helpful percent
* If viewers of that comment upvote the comment, then it adds to Helpfulness numerator.

* If viewers of that comment just make a comment, then it adds to Helpfulness denominator.

* **Note : you should filter out some cases with numerator > denominator**

* Plot the distribution of helpful percent (hint: .hist())

* **Ratio = -1, if both Helpfulness numerator & Helpfulness denominator are zero.**

```python
Helpful_percent = pd_data [[ "HelpfulnessNumerator", "HelpfulnessDenominator" ]]
Helpful_percent [ "Ratio" ] = Helpful_percent[ "HelpfulnessNumerator" ] / Helpful_percent[ "HelpfulnessDenominator" ]

Ratio = Helpful_percent [ "Ratio" ].fillna(-1)
Ratio = pd.DataFrame ( Ratio )

Ratio [ "Count" ] = 1
Ratio
Group = Ratio.groupby ( [ Ratio [ "Ratio" ]] )

Count = Group.Count.count()
Count = pd.DataFrame ( Count )
Count.reset_index ( inplace = True )
Count

plt.bar ( Count["Ratio"], Count["Count"], width = 0.3, align = "edge")
plt.grid ()
```

## Suicide Rates Overview 1985 to 2016
##### Compares socio-economic info with suicide rates by year and country

This compiled dataset pulled from four other datasets linked by time and place, and was built to find signals correlated to increased suicide rates among different cohorts globally, across the socio-economic spectrum.

Columns including : country, year, sex, age, suicides_no, population, suicides/100k pop, country-year, HDI for year（人類發展指數）, gdp_for_year（國內生產總值）, gdp_per_capita, generation

```python
pd_data = pd.read_csv ( "master.csv" )
pd_data
```

## 1. Top-10 Suicide Cases 
* Sort by number of suicides_no ( hint : "country" )
* Also, show its average suicides/100k pop

```python
pd_data = pd_data.rename ( columns = { "suicides/100k pop" : "suicides_100k_pop" } )
Group = pd_data.groupby ( [ pd_data [ "country" ]] )

Suicides_No = Group.suicides_no.sum()
Suicides_No = pd.DataFrame ( Suicides_No )
Suicides_Mean = Group.suicides_100k_pop.mean()
Suicides_Mean = pd.DataFrame ( Suicides_Mean )

Merge = pd.merge ( Suicides_No, Suicides_Mean, on = [ "country" ])
Merge

Result = Merge.sort_values ( ["suicides_no"], ascending = False )[:10]
Result.rename ( columns = { "suicides_no": "Suicides Number", "suicides_100k_pop": "Average Suicides / 100k Population" } )
```
